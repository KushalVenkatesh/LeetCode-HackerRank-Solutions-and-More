```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def st_paradox():
    pay_day = 0
    pay_day_list = []

    while len(pay_day_list) < 10000:
        while np.random.random() > .5:
            if pay_day == 0:
                pay_day += 2
            else:
                pay_day *= 2
        pay_day_list.append(pay_day)
        pay_day = 0
    return pay_day_list
```

```python
pay_day_list = st_paradox()

plt.hist(x = pay_day_list, bins = 16, lw = 2.5, edgecolor = 'Black')
plt.title('St. Petersburg Paradox Simulation (10,000 Runs)')
plt.ylabel('Observations')
plt.xlabel('$ Winning')
plt.rcParams.update({'font.size': 15})
plt.annotate(f'The average $ won per\n this simulation is ${np.mean(pay_day_list):.2f}', xy=(0.40, 0.75), xycoords='axes fraction');
```

```python
mean_pay_day = []

while len(mean_pay_day) < 100:
    mean_pay_day.append(np.mean(st_paradox()))
```

```python
plt.hist(x = mean_pay_day, bins = 10, lw = 2.5, edgecolor = 'Black')
plt.title('Average $ won (100 simulations)')
plt.ylabel('Observations')
plt.xlabel('Average $ Winning');
```

```python
print(f'The average of the 100 simulations of 10,000 games is {np.mean(mean_pay_day)}')
print(f'The highest average payout was: {max(mean_pay_day)}')
print(f'The lowest average payout was: {min(mean_pay_day)}')
```
